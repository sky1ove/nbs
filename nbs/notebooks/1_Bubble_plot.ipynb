```r
library(ggplot2)
```

```r
# Read data
data = read.csv('../data/bubble.csv')
```

```r
head(data); dim(data)
```

Data needs to be unstacked: x & y label columns (in this case, protein & cluster), and color and size as two value columns.

```r
# Plot
figsize = c(3,8)

options(repr.plot.width=figsize[1], repr.plot.height=figsize[2])
ggplot(data,aes(cluster,protein)) +
       geom_point(aes(size=size,color=color),shape=16)+
       facet_grid(Group ~., scales="free", space="free")+
       scale_colour_distiller(palette="RdYlBu")+
       labs(x="",y="") +
       # theme_pubr() + #theme_void() if vertical legend
       theme(strip.text.y = element_blank(),
             axis.line.x=element_blank(),
             axis.line.y=element_blank(),
             panel.background=element_rect(colour="black", fill="white"),
             panel.grid.major = element_blank(),  # remove major grid
             panel.grid.minor = element_blank(),  # remove minor grid
             axis.text.y=element_text(size=10),
             axis.text.x=element_text(angle=60,size=10,hjust=1))
```

```r
# Save
ggsave("Dotplot.pdf", width = figsize[1], height = figsize[2], useDingbats=F)
```
